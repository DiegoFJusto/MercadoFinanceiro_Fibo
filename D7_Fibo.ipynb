## Níveis de Fibo sob consulta ( dados Yahoo Finance )

importar as bibliotecas necessárias

```python
#!pip install yfinance
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

## Escolher o ativo (yahoofinance) e a data inicial

```python
ativo_in = str(input('Escolha seu ativo : [ ex: EURUSD=X] :'))
data_ini = str(input('Escolha data inicial : [ ex: 2020-01-01] :'))
ativo = yf.download(ativo_in, start=data_ini) 
```

## Mostrar em gráfico os preços de fechamento

```python
# Plot the close price on a chart
plt.figure(figsize=(21, 9))
plt.plot(ativo['Close'], color = 'red')
plt.title(ativo_in +' Index')
plt.xlabel('Date')
plt.ylabel('Points')
plt.show()
```

## Níveis requeridos: 0.236, 0.382 e 0.618 (0.5 incluir)

```python
max_point = ativo['Close'].max()
min_point = ativo['Close'].min()
diff = max_point - min_point
#print(f'Maximum: {max_point} \nMinimum: {min_point} \nDifference:{diff}')
```

```python
# Calculando os níveis
level1 = max_point - diff * 0.236
level2 = max_point - diff * 0.382
level3 = max_point - diff * 0.5
level4 = max_point - diff * 0.618
#print(f'(23.6%): {level1:.6} \n(38.2%): {level2:.6} \n(50%): {level3:.6} \n(61.8%): {level4:.6} \n')
```

```python
# Plotar Níveis de Fibonacci
plt.figure(figsize=(21, 9))
plt.title(ativo_in+' : Níveis de Retração de Fibonacci')
plt.plot(ativo.index, ativo['Close'], color = 'black')
plt.axhline(max_point, linestyle='-', alpha = 0.6, color='red')
plt.axhline(level1, linestyle='-', alpha = 0.8, color='maroon')
plt.axhline(level2, linestyle='-', alpha = 0.8, color='gold')
plt.axhline(level3, linestyle='-', alpha = 0.8, color='gold')
plt.axhline(level4, linestyle='-', alpha = 0.8, color='darkblue')
plt.axhline(min_point, linestyle='-', alpha = 0.6, color='blue')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
